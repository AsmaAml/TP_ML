```python
#Question 2

import numpy as np
import matplotlib.pyplot as plt

def compute_error_for_line_given_point(b, w, points):
    # Initialize total error to 0
    totalerror = 0
    
    # Loop over each point in the dataset
    for i in range(0, len(points)):
        # Extract the x and y coordinates of the point
        x = points[i, 0]
        y = points[i, 1]

        # Compute the squared difference between the actual y values and
        # the y values predicted by our line equation y = wx + b
        totalerror += (y - (w * x + b)) ** 2
    
    # Average the total error by the number of points, factor of 1/2 is
    # often used in the computation of mean squared error to simplify
    # the derivative (gradient) computation
    return totalerror / (2 * float(len(points)))

# Example usage:
# Create some sample data
points = np.array([[1, 2], [2, 4], [3, 6]])

# Define parameters of the line
b = 0  # y-intercept
w = 2  # slope

# Compute error
error = compute_error_for_line_given_point(b, w, points)
print("Computed Error:", error)
```

```python
#Question 3

def step_gradient(b_current, w_current, points, learning_rate):
    b_gradient = 0
    w_gradient = 0
    N = float(len(points))  # Number of points

    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]

        # Calculate the gradients of the loss function with respect to b and w
        b_gradient += (1 / N) * ((w_current * x + b_current) - y)  # Partial derivative with respect to b
        w_gradient += (1 / N) * x * ((w_current * x + b_current) - y)  # Partial derivative with respect to w

    # Update the values of b and w by taking a step proportional to the negative of the gradients
    new_b = b_current - (learning_rate * b_gradient)
    new_w = w_current - (learning_rate * w_gradient)

    return [new_b, new_w]
```

```python
#Question 4


import numpy as np
import matplotlib.pyplot as plt

def gradient_descent_runner(points, starting_b, starting_w, learning_rate, num_iterations):
    b = starting_b
    w = starting_w
    cost_history = np.zeros(num_iterations)  # Initialize an array to store cost history
    points = np.array(points)  # Ensure points are a NumPy array to improve performance

    for i in range(num_iterations):
        b, w = step_gradient(b, w, points, learning_rate)  # Update parameters
        cost = compute_error_for_line_given_point(b, w, points)  # Compute cost once per iteration
        cost_history[i] = cost  # Store the cost in history

        # Log current state
        print(f"Iteration {i + 1}:")
        print(f"Cost function = {cost:.4f}")
        print(f"w = {w:.4f}, b = {b:.4f}")

    # Plot the cost history over iterations
    plt.plot(range(num_iterations), cost_history)
    plt.title("Cost Function Across Iterations")
    plt.xlabel("Iteration")
    plt.ylabel("Cost")
    plt.show()

    return [b, w]

# Note: Make sure to define `step_gradient` and `compute_error_for_line_given_point` correctly.
```
