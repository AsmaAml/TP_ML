```python
import numpy as np  # Bibliothèque pour le calcul scientifique et la manipulation de tableaux en Python
import matplotlib.pyplot as plt  # Module pour la création de visualisations graphiques en Python
from sklearn.datasets import make_regression  # Fonction pour générer un jeu de données de régression pour les tests
from sklearn.linear_model import SGDRegressor  # Importe SGDRegressor, un modèle de régression linéaire utilisant le gradient stochastique descendant
```

```python
np.random.seed(0)  # Initialise le générateur de nombres aléatoires avec une graine fixe pour la reproductibilité
# Générer un jeu de données de régression avec 100 échantillons, 1 caractéristique et du bruit ajouté
x, y = make_regression(n_samples=100, n_features=1, noise=10)
```

```python
plt.scatter(x, y)  # Afficher un graphique de dispersion des points de données (x, y)
```

```python
# Créer un modèle de régression linéaire utilisant la descente de gradient stochastique (SGD)
model = SGDRegressor(max_iter=100, eta0=0.0001)
```

```python
model.fit(x, y)  # Entraîner le modèle de régression linéaire sur les données (x, y)
```

```python
# Afficher le coefficient de détermination du modèle sur les données d'entraînement
print("coefficient de détermination= ", model.score(x, y))
# Afficher un graphique de dispersion pour visualiser les points de données (x, y)
plt.scatter(x, y)
```

```python
plt.scatter(x, y)  # Afficher un graphique de dispersion des points de données (x, y)
# Tracer la ligne de prédiction du modèle sur le graphique de dispersion, en jaune et avec une épaisseur de ligne de 3
plt.plot(x, model.predict(x), c='yellow', lw=3)
```

```python
model = SGDRegressor(max_iter=1000, eta0=0.001)
model.fit(x,y)
print("coefficient de détermination= ",model.score(x,y))
plt.scatter(x,y)
plt.plot(x, model.predict(x), c='yellow', lw=3)
```
